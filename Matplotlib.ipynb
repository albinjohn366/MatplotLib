# Matplotlib
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
```

```
# Drawing a line 
x_points = np.array([0, 6])
y_points = np.array([0, 250])
plt.plot(x_points, y_points)
plt.show()
```

```
# Drawing a graph 
plt.plot([2, 5, 6, 7], [1, 3, 9, 2])
plt.show()
```

```
# To plot only points
plt.plot([2, 5, 6, 7], [1, 3, 9, 2], 'o')
plt.show()
```

```
# Default x point
plt.plot([2, 5, 6, 3, 4, 1, 10, 2])
plt.show()
```

```
# Formatting line and dots
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], 'o:r')
plt.show()
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], '*-k')
plt.show()
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], 'o-.c', ms=20)
plt.show()
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], 'o-.c', ms=20, mec='r')
plt.show()
```

```
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], 'o-.c', ms=20, mfc='k')
plt.show()
```

```
# TO change the color of the lines
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], color='#8153E5')
plt.show()
# TO change the width of the lines
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], color='#8153E5', linewidth=10)
plt.show()
```

```
# Drawing multiple lines in the same graph
# TO change the color of the lines
plt.plot([2, 5, 6, 3, 4, 1, 10, 2], color='#8153E5')
plt.plot([1, 5, 2.4, 5, 1, 4], color='r')
plt.plot([5, 5, 1.4, 10, 4, 2], 'o',  color='c', ls=':', mec='y', mfc='k', ms=10)
plt.show()
```

```
# One can also plot many lines using one plot function
plt.plot([1, 5, 24], [2, 3, 5], [2.4, 4, 12], [5, 8, 90])
plt.show()
```

```
# Labelling
years = np.array([2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012])
price = np.array([7, 9, 10, 10.5, 9, 10, 11, 15])
font1 = {'family': 'serif', 'size': 10, 'color': 'k'}

plt.plot(years, price, color='r')
plt.title("Price of the product over years", loc='left', fontdict=font1)
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Price', fontdict=font1)
plt.show()
```

```
# Grid
years = np.array([2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012])
price = np.array([7, 9, 11, 15, 9, 10, 21, 5])
font1 = {'family': 'serif', 'size': 10, 'color': 'k'}

plt.plot(years, price, color='r')
plt.title("Price of the product over years", loc='left', fontdict=font1)
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Price', fontdict=font1)
plt.grid()
plt.show()

plt.plot(years, price, color='r')
plt.title("Price of the product over years", loc='left', fontdict=font1)
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Price', fontdict=font1)
plt.grid(axis='x')
plt.show()

plt.plot(years, price, color='r')
plt.title("Price of the product over years", loc='left', fontdict=font1)
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Price', fontdict=font1)
plt.grid(axis='y')
plt.show()

plt.plot(years, price, color='r')
plt.title("Price of the product over years", loc='left', fontdict=font1)
plt.xlabel('Years', fontdict=font1)
plt.ylabel('Price', fontdict=font1)
plt.grid(color='g', ls=':')
plt.show()
```

```
# Mutple figures of plot, individual titles and super title

# Plot 1
y = [1, 2, 4, 5]
plt.subplot(2, 3, 1)
plt.title("Plot 1")
plt.plot(y)

# Plot 2
y = [4, 2, 1]
plt.subplot(2, 3, 2)
plt.title("Plot 2")
plt.plot(y)

# Plot 3
y = [4, 9, 3]
plt.subplot(2, 3, 3)
plt.title("Plot 3")
plt.plot(y)

# Plot 4
y = [4, 4, 4]
plt.subplot(2, 3, 4)
plt.title("Plot 4")
plt.plot(y)

# Plot 5
y = [40, 14, 34]
plt.subplot(2, 3, 5)
plt.title("Plot 5")
plt.plot(y)

# Plot 6
y = [14, 24, 34]
plt.subplot(2, 3, 6)
plt.title("Plot 6")
plt.plot(y)

plt.suptitle("All Plots")
plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

```
# Scatter Plots
speed = [20, 40, 30, 100, 45, 39]
age = [30, 20, 20, 1, 5, 2]
plt.scatter(age, speed, color='k', s=100)

speed = [30, 0, 40, 120, 35, 49]
age = [30, 20, 20, 1, 5, 2]
color = ['black', 'orange', 'yellow', 'red', 'cyan', 'pink']
plt.scatter(age, speed, color=color)
plt.show()
```

```
# Bar Graph
groups = ["Red", "Yellow", "Green", "Black"]
scores = [200, 350, 100, 400]
plt.bar(groups, scores)
plt.show()
```

```
# Horizontal Bar
groups = ["Red", "Yellow", "Green", "Black"]
scores = [200, 350, 100, 400]
plt.barh(groups, scores)
plt.show()
```

```
# Bar graph with color
groups = ["Red", "Yellow", "Green", "Black"]
scores = [200, 350, 100, 400]
plt.bar(groups, scores, color='r')
plt.show()
```

```
# Bar graph with different colors
groups = ["Red", "Yellow", "Green", "Black"]
scores = [200, 350, 100, 400]
colors = ['r', 'y', 'g', 'k']
plt.bar(groups, scores, color=colors, width=0.5)

for i, value in enumerate(scores):
  plt.text(i - 0.1, value + 5, str(value))

plt.grid(color='b', ls='-.', linewidth=0.2)
plt.show()
```

```
# Plotting Histoagram
x = np.random.normal(120, 10, 250)
plt.hist(x)
plt.show()
```

```
# Plotting in Pie chart
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
plt.pie(x, labels=labels)
plt.show()
```

```
# Changing start angle
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
plt.pie(x, labels=labels, startangle=60)
plt.show()
```

```
# Explode cells
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
explodes = [0.1, 0, 0, 0]
plt.pie(x, labels=labels, startangle=60, explode=explodes)
plt.show()
```

```
# Having Shadows
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
explodes = [0.1, 0, 0, 0]
shadows = [0.2, 0, 0, 0]
plt.pie(x, labels=labels, startangle=60, explode=explodes, shadow=shadows)
plt.show()
```

```
# Different colors
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
colors = ['b', 'y', 'k', 'r']
explodes = [0.1, 0, 0, 0]
shadows = [0.2, 0, 0, 0]
plt.pie(x, labels=labels, startangle=60, explode=explodes, shadow=shadows, colors=colors)
plt.legend()
plt.show()
```

```
# To represent percentage
x = [24, 56, 12, 54]
labels = ["Science", "English", "Maths", "Biology"]
colors = ['b', 'y', 'k', 'r']
explodes = [0.1, 0, 0, 0]
shadows = [0.2, 0, 0, 0]
patches, texts, auto_texts = plt.pie(x, labels=labels, startangle=60, explode=explodes, shadow=shadows, colors=colors, autopct='%1.1f%%')
for autotext in auto_texts:
    autotext.set_color('white')
plt.legend()
plt.show()
```
